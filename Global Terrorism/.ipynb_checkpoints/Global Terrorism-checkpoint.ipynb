<p>Only selecting necessary columns</p>

```python
df = df[['iyear', 'imonth', 'iday', 'country_txt', 'region_txt', 'city', 'latitude', 'longitude', 'attacktype1_txt', 'targtype1_txt', 'target1', 'gname', 'weaptype1_txt', 'nkill', 'nwound']]
df
```

<p>Renaming some columns

```python
df = df.rename(columns={'iyear': 'year', 'imonth': 'month', 'iday': 'day', 'country_txt': 'countryName', 'region_txt': 'regionName', 'attacktype1_txt': 'attackType', 'targtype1_txt': 'target', 'target1': 'specific_target', 'gname': 'group', 'weaptype1_txt': 'weaponType', 'nkill': 'killed', 'nwound': 'wounded'})
df
```

<h3>Data Visualisation</h3>

Number of terrorist attacks every year

```python
plt.style.use('seaborn')
plt.subplots(figsize=(20,6))
sns.countplot('year', data=df, palette='rocket')
plt.title('Number of terrrorist attacks each year')
plt.show()
```

- - -
Bar Plot of number of people killed and wounded due to terrorist attacks every year

```python
cas = df[['year', 'killed', 'wounded']]
cas = cas.groupby('year').sum().reset_index()

cas.plot(x='year', y=['killed', 'wounded'], kind='bar', figsize=(20,8))
plt.title('Killed and wounded every year')
plt.show()
print("This graph shows the number of people killed(blue) or wounded(green) every year due to terrorist attacks. The graph has been going up ever since the late 1970s apart from a dip during the late 1990s and early 2000s. ")
```

- - -
Countries that have faced the maximum number of attacks

```python
plt.style.use('seaborn')
plt.subplots(figsize=(20,6))
atck_country = df.countryName.value_counts()[:20].reset_index()

atck_country.columns = ['countryName', 'totalAttacks']
atck_country
sns.barplot(x=atck_country.countryName, y=atck_country.totalAttacks)
plt.title('Number of terrrorist attacks country wise')
plt.show()
print("As is evident from the bar plot, Iraq has the most number of terrorist attacks since 1970 followed by Pakistan, Afghanistan and India.")
```

- - -
Region wise

```python
plt.style.use('seaborn')
plt.subplots(figsize=(20,7))
atck_region = df.regionName.value_counts().reset_index()
atck_region.columns = ['regionName', 'totalAttacks']
sns.barplot(x=atck_region.regionName, y=atck_region.totalAttacks)
plt.title('Region wise terrorist attacks')
plt.show()
print("Middle East, North Africa and South Asia are the regions that are most affected by terrorism")
```

- - - 
Loacations of terrorist attacks

```python
plt.subplots(figsize=(20,10))
sc = sns.scatterplot(data=df, x='longitude', y='latitude').set(xlim=(-180,180), ylim=(-90,90))

plt.title('Locations of terrorist attacks according to latitude and longitude')
plt.show()
print("This plot shows all the locations of terrorist attacks included in the dataset")
```

- - -
Most used weapons in terrorist attacks

```python
plt.style.use('seaborn')
plt.subplots(figsize=(10,7))
atck_weaptype = df.weaponType.value_counts().reset_index()
atck_weaptype.columns = ['weaponType', 'totalAttacks']
sns.barplot(x=atck_weaptype.weaponType, y=atck_weaptype.totalAttacks, color='orange')
plt.xticks(rotation=90)
plt.title('Weapon type wise terrorist attacks')
plt.show()
```

- - -
Known groups

```python
plt.style.use('seaborn')
plt.subplots(figsize=(20,7))
atck_group = df.group.value_counts()[1:15].reset_index()
atck_group.columns = ['group', 'totalAttacks']

sns.barplot(x=atck_group.group, y=atck_group.totalAttacks)
plt.xticks(rotation=90)
plt.title('Group wise terrorist attacks')
plt.show()
```

```python
sns.pairplot(df, vars=['year', 'month', 'killed', 'wounded'])
plt.show()
```

- - -
Major targets

```python
plt.style.use('seaborn')
plt.subplots(figsize=(20,7))
atck_target = df.target.value_counts()[0:15].reset_index()
atck_target.columns = ['target', 'totalAttacks']

sns.barplot(x=atck_target.target, y=atck_target.totalAttacks)
plt.xticks(rotation=90)
plt.title('Targets of terrorist attacks')
plt.show()
```

<p>From these plots, we can arrive at a few conclusions.</p>
<p>The Middle East is historically the region most affected by terrorist activities, that includes Iraq which has had the most number of terrorist attacks since 1970.</p>
<p>Number of terrorist attacks have continually increased since about 1980 with a dip around the turn of the millenium. They have shot up in the last 10 years or so.</p>
<p>Most of the attacks have been carried out with the use of explosives and firearms.</p>
<p>Terrorist groups have carried out attacks targeting civilian lives and property. Military, police, government and business have been the next biggest targets(among those targets that are known).
